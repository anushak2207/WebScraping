```python
jobs=driver.find_elements_by_xpath('//article[@class="jobTuple bgWhite br4 mb-8"]') # Main tag that contains all the job related tags

data={}
df=pd.DataFrame(columns=['Title','Company', "Exp", 'Salary', 'Location'])

index=0
for job in jobs:
    for j in job.find_elements_by_xpath('div[@class="jobTupleHeader"]'):     #sub division where all the job related elements are present
        data["Title"]=j.find_element_by_xpath('div[@class="info fleft"]').text  #Fetching the job title from each job
        data["Company"]=j.find_element_by_xpath('div/div/a').text               #Fetching the Company Name from each job
        data["Exp"]=j.find_element_by_xpath('div/ul/li[1]').text                #Fetching the Experience from each job
        data["Salary"]=j.find_element_by_xpath('div/ul/li[2]').text             #Fetching the Salary details from each job
        data["Location"]=j.find_element_by_xpath('div/ul/li[3]').text           #Fetching the Location details from each job
       
        
        df=df.append(data, ignore_index=True)                                   #Storing in dataframe by Appending all the data one by one
        index+=1                                                                #Count/iteration
print("Total Jobs:", index)
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
```

```python
#Activating the Chrome Browser
driver=webdriver.Chrome('C:/Users/ak38/Desktop/FlipDataStore/chromedriver')
driver.get("https://www.flipkart.com/")      #Opening the Home page of flipkart
```

```python
product_search=driver.find_element_by_name("q")        #Locating the product search bar by class name
product_search.clear()        #Clearing the search bar to make sure there is no history search
product_search.send_keys("Sunglasses")        #Searching the product
Search_button=driver.find_element_by_class_name("L0Z3Pu")  #Locating search button
Search_button.click()     #Click
```

```python
data={}                      #Dictionary to store data

df=pd.DataFrame(columns=['Brand', 'Prod_desc', 'Price', 'Discount'])   #Creating a dataframe with columns to store data

i=1
j=0


while i<=3:
    glasses=driver.find_elements_by_xpath('//div[@class="_1AtVbE col-12-12"]/div/div') # Main tag
    for glass in glasses:
        for g in glass.find_elements_by_xpath('div[@class="_1xHGtK _373qXS"]'):     #sub division where all the glass related elements are present
            data["Brand"]=g.find_element_by_class_name('_2WkVRV').text  #Fetching the brand from each sunglass
            data["Prod_desc"]=g.find_element_by_class_name('IRpwTa').text               #Fetching the Description from each sunglass
            data["Price"]=g.find_element_by_xpath('div/a/div/div').text                #Fetching the Price from each sunglass
            try:
                data["Discount"]=g.find_element_by_xpath('div/a/div/div/span').text             #Fetching the discount details from each sunglasses        
            except NoSuchElementException:
                data['Discount']="Null"                           #If discount not found, fill in as NULL
            df=df.append(data, ignore_index=True)                   #storing the data
            j+=1
        if j==100:
            break

       
        
    try:
        next =  driver.find_element_by_xpath('//div[@class="_1AtVbE col-12-12"]/div/div/nav/a[12]/span').click()   #going to next page
        time.sleep(4)
    except NoSuchElementException:
        next =  driver.find_element_by_xpath('//div[@class="_1AtVbE col-12-12"]/div/div/nav/a/span').click()
        time.sleep(4)
        
    i=i+1
           
print(df)
```

```python
data={}                      #Dictionary to store data

df=pd.DataFrame(columns=['Title', 'Rating', 'Price'])   #Creating a dataframe with columns to store data


laptops=driver.find_elements_by_xpath('//div[@class="a-section a-spacing-medium"]/div[2]/div[2]') # Main tag
for laptop in laptops[:10]:
    for s in laptop.find_elements_by_xpath('div[@class="sg-col-inner"]'):     #sub division where all the product related elements are present
        data["Title"]=s.find_element_by_xpath('div[@class="a-section a-spacing-none"]/h2/a/span').text  #Fetching the brand from each product
        try:
            data["Rating"]=s.find_element_by_xpath('div[@class="a-section a-spacing-none a-spacing-top-micro"]/div/span').get_attribute('aria-label')
        except NoSuchElementException:
            data["Rating"]="Null"
        
        try:
            data["Price"]=s.find_element_by_xpath('div[@class="sg-row"]/div/div/div/div/a/span/span/span[2]').text                #Fetching the Price from each product
        except NoSuchElementException:
            data["Price"]="Null"
            
        df=df.append(data, ignore_index=True) 
df
```

###### Assignment 4.1:
    Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
 You need to find following details:
 
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#Activating the Chrome Browser
driver=webdriver.Chrome('C:/Users/ak38/Desktop/FlipDataStore/chromedriver')

#Opening the URL
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")      

# to maximize the browser window
driver.maximize_window()
```

```python
#Extracting the table content from "Top 30 most-viewed YouTube videos"
videos=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr')

len(videos)
```

```python
#Creating the lists to store the extracted data
vrank=[]
vname=[]
vuploader=[]
vdate=[]
vviews=[]

#Fetching the ranks from the "Top 30 most-viewed YouTube videos" table
rank=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for r in rank:
    vrank.append(r.text.replace(".",""))

#Fetching the video names from the "Top 30 most-viewed YouTube videos" table
name=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for n in name:
    vname.append(n.text.replace('"','').split("[")[0])

#Fetching the video uploader/artist name from the "Top 30 most-viewed YouTube videos" table
uploader=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for u in uploader:
    vuploader.append(u.text)

#Fetching the upload date from the "Top 30 most-viewed YouTube videos" table
date=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for d in date:
    vdate.append(d.text)

#Fetching the no.of views(in billions) from the "Top 30 most-viewed YouTube videos" table
views=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for v in views:
    vviews.append(v.text)
```

```python
#Creating a DataFrame
data=pd.DataFrame({})
#Assigning all the lists to the dictionary keys
data['Rank']=vrank
data['Name']=vname
data['Artist']=vuploader
data['Upload_Date']=vdate
data['Views (in billions)']=vviews
#viewing the final dataset
data
```

```python
#Creating the lists to append the data points
ds_name=[]
dtype=[]
task=[]
attr_type=[]
instances=[]
attributes=[]
year=[]

#Main elements from which the data elements are extracted
table=driver.find_elements_by_xpath('//table/tbody/tr/td[2]/table[2]/tbody/tr')

#Iterating through the main element to extract the data points
for tab in table:
    #Fetching the data set name
    try:
        ds_name.append(tab.find_element_by_xpath('td[1]/table/tbody/tr/td[2]/p/b').text)
    except NoSuchElementException:
        ds_name.append('-')
    
    #Fetching the data type
    try:
        dtype.append(tab.find_element_by_xpath('td[2]/p').text)
    except NoSuchElementException:
        dtype.append('-')
    
    #Fetching the task
    try:
        task.append(tab.find_element_by_xpath('td[3]/p').text)
    except NoSuchElementException:
        task.append('-')
        
    try:
        attr_type.append(tab.find_element_by_xpath('td[4]/p').text)
    except NoSuchElementException:
        attr_type.append('-')
        
    try:
        instances.append(tab.find_element_by_xpath('td[5]/p').text)
    except NoSuchElementException:
        instances.append('-')
        
    try:
        attributes.append(tab.find_element_by_xpath('td[6]/p').text)
    except NoSuchElementException:
        attributes.append('-')
        
    try:
        year.append(tab.find_element_by_xpath('td[7]/p').text)
    except NoSuchElementException:
        year.append('-')

#Verifying if all the entries are fetched
len(ds_name), len(dtype), len(task), len(attr_type), len(instances), len(attributes), len(year)
```

```python
#Creating the dataframe
data=pd.DataFrame({})

#Appending each list to the dictionary 
data["DataSet_Name"]=ds_name
data["Data_Type"]=dtype
data["Task"]=task
data["Attribute_Type"]=attr_type
data["Instances"]=instances
data["Attributes"]=attributes
data["Year"]=year


pd.set_option('display.max_rows', None)   #this code helps to view all the rows without hiding them

#Viewing the final data
data
```

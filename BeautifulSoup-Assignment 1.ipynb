```python
print(soup.prettify())
```

###### Assignment 1.2:
    Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
response=requests.get("https://www.imdb.com/chart/top?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=cb6cf75a-1a51-49d1-af63-8202cfc3fb01&pf_rd_r=KVT9E6420VD4ZXK5J53G&pf_rd_s=right-4&pf_rd_t=15506&pf_rd_i=moviemeter&ref_=chtmvm_ql_3")
response
```

```python
data=response.text
```

```python
soup=BeautifulSoup(data, 'html.parser')
soup
```

```python
movies=soup.find_all('td', class_='titleColumn')                 #Movie Name
year=soup.find_all('span', class_='secondaryInfo')               #Year of Release
rating=soup.find_all('td', class_='ratingColumn imdbRating')     #IMDB Rating
```

```python
#Lets extract all the top rated movies
MovieName=[]
YearOfRelease=[]
IMDBrating=[]

for movie in movies:
    MovieName.append(movie.text[16:-8].strip())
for y in year:
    YearOfRelease.append(y.text[1:5])        #=soup.find('span', class_='secondaryInfo').text
for rate in rating:
    IMDBrating.append(rate.text[1:4])       #=soup.find('td', class_='ratingColumn imdbRating').text
    
print(MovieName, YearOfRelease, IMDBrating)
```

```python
#Lets make it dataframe for better visualization
import pandas as pd
data=pd.DataFrame({})
data['MovieName']=MovieName
data['YearOfRelease']=YearOfRelease
data['IMDBrating']=IMDBrating

```

```python
data.sort_values(ascending=False, by='IMDBrating', inplace=True)
pd.set_option('display.max_rows', None) 
data.head(100)#Top 100 movies sorted by IMDB Rating
```

```python
#Saving the result as CSV file

data.to_csv("Top Rated Movies.csv")
data.head(100).to_csv("Top Rated 100 Movies.csv")
```

###### Assignment 1.4:
    Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’
  

```python
r=requests.get("https://bookpage.com/reviews")
r
```

```python
data=r.text
data
```

```python
soup=BeautifulSoup(data, 'html.parser')
soup
```

```python
Books=soup.find_all('div', class_='flex-article-content')
Authors=soup.find_all('p', class_='sans bold')
Genre=soup.find_all('p', class_='genre-links hidden-phone')
Reviews=soup.find_all('p', class_='excerpt')
```

```python
BookName=[]
AuthorName=[]
Genres=[]
BookReviews=[]

for book in Books:   
    BookName.append(book.text[2:].split('\n')[0])
for author in Authors:
    AuthorName.append(author.text.replace('\n', ''))
for genre in Genre:
    Genres.append(genre.text.replace('\n', ''))
for review in Reviews:
    BookReviews.append(review.text.replace('\n', ''))

print(BookName,'\n','\n', AuthorName, '\n','\n', Genres, '\n','\n', BookReviews )
```

```python
#Lets make it dataframe for better visualization
import pandas as pd
data=pd.DataFrame({})
data['BookName']=BookName
data['AuthorName']=AuthorName
data['Genres']=Genres
data['BookReviews']=BookReviews
```

```python
data
```

```python
#Saving the result as CSV file

data.to_csv("BookReviews.csv")
```

    Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
    i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 
    ii) Top 10 ODI Batsmen in men along with the records of their team and rating. 
    iii)Top 10 ODI bowlers along with the records of their team and rating.


###### i) 
    Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

```python
#importing the required Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
r=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
r
```

```python
data=r.text
```

```python
soup=BeautifulSoup(data, 'html.parser')
```

```python
#Create a dictionary to save the data

d = {'key': 'value'}
d
```

```python
tables=soup.find_all('tr', class_='table-body')
tables
```

```python
ODITable = {}
index = 0

for table in tables:
    team = table.find('span', class_='u-hide-phablet').text
    match = table.find('td', class_='table-body__cell u-center-text').text
    point = table.select('td')[3].text
    rating = table.find('td', class_='table-body__cell u-text-right rating').text
    index+=1
    
    ODITable[index] = [team, match, point, rating]
    
    
    
    print("Teams:", team, "\nMatches:", match, "\nPoints:", point, "\nRating:", rating, '\n-----')
    
print("Total Teams: ", index)
```

```python
ODITable_df=pd.DataFrame.from_dict(ODITable, orient = 'index', columns= ["Teams", "Matches", "Points", "Rating"])
```

```python
#Top 10 ODI teams in men’s cricket
ODITable_df.head(10)
```

###### Assignment 1.7:
    Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.


```python
#importing the required Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
amazon=requests.get("https://www.amazon.in/s?k=phones+under+20000&ref=nb_sb_noss_2")
amazon
```

```python
data=amazon.text
```

```python
soup=BeautifulSoup(data, 'html.parser')
```

```python
phones=soup.find_all('div', class_='s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col sg-col-12-of-16')
phones
```

```python
PhoneTable = {}
index = 0

for cell in phones:
    cellName = cell.find('a', class_='a-link-normal a-text-normal').text
    price=cell.find('span', class_='a-price-whole').text
    rating=cell.find('i').text
    imageLink=cell.find('div', class_='a-section aok-relative s-image-fixed-height').find('img').get('src')

    index+=1
    
    PhoneTable[index] = [cellName, price, rating, imageLink]
    
    print("Phones:", cellName, "\nPrice:", price, "\nRatings:", rating, "\nImageURL", imageLink, '\n---')
print("Total Phones:", index)
```

```python
PhoneTable_df=pd.DataFrame.from_dict(PhoneTable, orient = 'index', columns= ["Phones", "Price", "Ratings", "ImageURL"])
```

```python
PhoneTable_df.sort_values(by="Ratings", ascending=False, inplace=True)
```

```python
#all the mobile phones under Rs. 20,000 listed on Amazon.in
PhoneTable_df
```

###### Assignment 1.8: 
    Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
#importing the required Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
response=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJ_4rXnitnI")
response
```

```python
data=response.text
```

```python
soup=BeautifulSoup(data, 'html.parser')
```

```python
weather=soup.find_all('li', class_='forecast-tombstone')
weather
```

```python
WeatherTable = {}
index = 0

for sunny in weather:
    period = sunny.find('p', class_='period-name').text
    shdisc = sunny.find('p', class_='short-desc').text
    temp = sunny.select('p')[3].text

    index+=1
    
    WeatherTable[index] = [period, shdisc, temp]
    
    print("Periods:", period, "\nshort_disc:", shdisc, "\nTemp:", temp, '\n---')
print("Total forecasts:", index)
```

```python
WeatherTable_df=pd.DataFrame.from_dict(WeatherTable, orient = 'index', columns= ["Periods", "short_disc", "Temp"])
```
